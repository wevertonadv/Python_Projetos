```python
import pandas as pd

# Carrega os dados do Excel
df = pd.read_excel('tabela_vivo.xlsx', sheet_name='devices')

# Exibe as primeiras linhas para verificação
df.head()

# Cria uma estrutura de dados aninhada
categorias = {}
for _, row in df.iterrows():
    categoria = row['CATEGORIA'].title()
    vertical = row['VERTICAL'].title()
    subcategoria = row['SUBCATEGORIA_GERENCIAL'].title()

    if categoria not in categorias:
        categorias[categoria] = {}
    if vertical not in categorias[categoria]:
        categorias[categoria][vertical] = []

    if subcategoria not in categorias[categoria][vertical]:
        categorias[categoria][vertical].append(subcategoria)

def gerar_html_filtro(categorias):
    svg_icon = (
        '<svg class="toggle-icon h-4 w-4" xmlns="http://www.w3.org/2000/svg" viewBox="0 0 20 20" fill="currentColor">'
        '<path fill-rule="evenodd" d="M5.293 7.293a1 1 0 011.414 0L10 10.586l3.293-3.293a1 1 0 111.414 1.414l-4 4a1 1 0 01-1.414 0l-4-4a1 1 0 010-1.414z" clip-rule="evenodd" />'
        '</svg>'
    )

    html = '<div class="mb-4">\n'
    html += '  <button onclick="toggleCategories(this)" class="toggle-categories">\n'
    html += '    Categorias\n'
    html += f'    {svg_icon}\n'
    html += '  </button>\n'
    html += '  <div class="categoriesContent bg-white border border-t-0 rounded-b hidden">\n'
    html += '    <div class="p-4">\n'

    for categoria, verticais in categorias.items():
        categoria_id = categoria.replace(" ", "-").lower()
        html += f'    <div class="mb-2 categoriaItem">\n'
        html += f'      <button onclick="toggleVertical(\'{categoria_id}\')" class="toggle-categories">\n'
        html += f'        {categoria}\n'
        html += f'        {svg_icon}\n'
        html += '      </button>\n'
        html += f'      <div class="categoriaContent hidden" id="{categoria_id}">\n'
        for vertical, subcategorias in verticais.items():
            vertical_id = f"{categoria}-{vertical}".replace(" ", "-").lower()
            html += '      <div class="mb-2 verticalContainer">\n'
            html += f'        <button onclick="toggleVertical(\'{vertical_id}\')" class="toggle-categories">\n'
            html += f'          {vertical}\n'
            html += f'          {svg_icon}\n'
            html += '        </button>\n'
            html += f'        <div class="verticalContent hidden" id="{vertical_id}">\n'
            html += f'          <label class="subcategoriesSelectAll">\n'
            html += f'            <input type="checkbox" class="subcategories" onclick="selectAll(this, \'{vertical_id}\')">\n'
            html += '            <span class="ml-2">Selecionar todos</span>\n'
            html += '          </label>\n'

            for subcategoria in subcategorias:
                html += '          <label class="inline-flex items-center">\n'
                html += '            <input type="checkbox" class="subcategories">\n'
                html += f'            <span class="ml-2">{subcategoria}</span>\n'
                html += '          </label>\n'

            html += '        </div>\n'
            html += '      </div>\n'
        html += '    </div>\n'

    html += '    </div>\n'
    html += '  </div>\n'
    html += '</div>\n'
    return html

# Gerar HTML dos filtros
html_filtros = gerar_html_filtro(categorias)

print(html_filtros)



print(html_filtros)
```
